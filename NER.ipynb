```python
patent_trade = pd.read_csv('/content/drive/My Drive/patent_trade_dataset.csv')
```

```python
patent_trade_dt = patent_trade[['or_name']]
patent_trade_dt = patent_trade_dt.drop_duplicates()
#patent_trade_dt['ee_name'] = patent_trade_dt['ee_name'].str.split(",").str[0]
patent_trade_dt = patent_trade_dt[:50]
patent_trade_dt.head()
```

```python
# use deeppavlov


from deeppavlov import build_model

ner_model = build_model('ner_ontonotes_bert_mult', download=True)


# define a function to extract named entities and identify their labels
def extract_entities(text):
    doc = ner_model(text)
    return doc[1][0][0]



# apply the extract_entities function to the "ee_name" and "or_name" columns
patent_trade_dt['or_name_reco'] = patent_trade_dt['or_name'].apply(lambda x: [[x]])

patent_trade_dt['or_name_entities'] = patent_trade_dt['or_name_reco'].apply(lambda x: extract_entities(x))
#patent_trade_dt['or_name_entities'] = patent_trade_dt['or_name'].apply(extract_entities)

patent_trade_dt
```

```python
#use spacy

#spacy.prefer_gpu()
nlp = spacy.load('en_core_web_sm')

# define a function to extract named entities and identify their labels
def extract_entities(text):
    doc = nlp(text)
    for ent in doc.ents:
      return ent.label_


patent_trade_dt['or_name_reco'] = patent_trade_dt['or_name'].apply(lambda x: [[x]])

patent_trade_dt['or_name_entities'] = patent_trade_dt['or_name'].apply(extract_entities)

patent_trade_dt
```
